```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
try:
    source =requests.get("https://www.doctorsfinder.in/doctors/mumbai/")
    source.raise_for_status()
    soup = BeautifulSoup(source.text, 'html.parser')
   
    hs= soup.find('div',class_  ="tab-pane active").find_all('h4')
    
    hr= soup.find_all('div',class_  ="card-body college-card")
    
   
        
    aa=[]
    bb=[]
    
    for i in hs[:50]:
        y= i.text.strip()
        aa.append(y)
        
    for j in hr[:50]:
        l=j.find('p').text.strip()
        bb.append(l)
        
    print(aa)
    print(len(aa))   
    print(bb)
    print(len(aa))

except Exception as e:
    print(e)

city ='Mumbai'
cit=[]
cit.extend([city for i in range(50)])


ds1={'Doctor_Name':aa,
    'feild':bb,
     'city':cit
    }
pd1= pd.DataFrame(ds1)
print(pd1)
```
