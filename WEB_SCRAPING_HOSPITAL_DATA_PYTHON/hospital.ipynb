```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
try:
    source =requests.get("https://www.policyx.com/health-insurance/network-hospitals/care-hyderabad-telangana/")
    source.raise_for_status()
    soup = BeautifulSoup(source.text, 'html.parser')
    hs= soup.find('table',class_  ="table table-border mt-3 text-center table-c").find_all('tr')
   
    a=[]
    b=[]
    c=[]
    d=[]
    e=[]
    
    for i in hs[:50]:
        x= i.find_all('td')[1].text
        a.append(x)
        y= i.find_all('td')[2].text
        b.append(y)
        z= i.find_all('td')[6].text
        c.append(z)
        p= i.find_all('td')[5].text
        d.append(p)
        w= i.find_all('td')[4].text
        e.append(w)
        
    print(a)
    print()
    print(b)
    print()
    print(c)
    print()
    print(d)
    print()
    print(e)
    print()
    print(len(a))
    
except Exception as e:
    print(e)
```

```python
mydataset = {a[0] : a[1:],
             b[0] : b[1:],
             c[0] : c[1:],
             d[0] : d[1:],
             e[0] : e[1:]
             
            }
print(mydataset)
```

```python
myvar = pd.DataFrame(mydataset)
myvar
```
